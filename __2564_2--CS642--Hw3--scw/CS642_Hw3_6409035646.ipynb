```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the input image
img = cv2.imread('CsAtTu.bmp')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# determine the kernel
kernel = np.ones((3,3), np.uint8)
```

# ก

## The most basic morphological operations are two: Erosion and Dilation

```python
img_erosion = cv2.erode(img, kernel, iterations=1) #ใช้ Erosion 
img_dilation = cv2.dilate(img, kernel, iterations=3) #ใช้ Dilation
```

```python
figure = plt.figure(figsize=(7,7))
plt.axis('off');
plt.imshow(img_erosion,cmap = 'gray')
```

```python
import imutils
from imutils.contours import sort_contours
from matplotlib import cm
from imutils import contours
```

```python
# img_dilation = cv2.dilate(img, kernel, iterations=1)
```

```python
# preprocess with canny to make it binary
edged = cv2.Canny(img, 30, 250) 
```

```python
figure = plt.figure(figsize=(7,7))
plt.axis('off');
plt.imshow(edged,cmap=cm.binary_r);
```

```python
# find contours
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts = imutils.grab_contours(cnts)
cnts, _ = contours.sort_contours(cnts, method="left-to-right")
```

```python
print("amount of objects detected = ",len(cnts))
```

```python
figure = plt.figure(figsize=(7,7))
plt.axis('off');
plt.imshow(cv2.drawContours(img.copy(), cnts, -1, (190,120,149), 5))
```

```python
# After finding contours, sorting from left-to-right, then croping

OBJ_number = 0
for c in cnts:
    area = cv2.contourArea(c)
    if area > 30: # determine threshold to control which is exact character contour (seeing exported images)
       
        # create square box for export croped character from the image
        x,y,w,h = cv2.boundingRect(c)
        # croping
        OBJ = 255 - img[y:y+h, x:x+w]
        # export croped image
        cv2.imwrite('OBJ_{}.png'.format(OBJ_number), OBJ)
        
        # fine majorAxis, minorAxis
        (x, y), (minorAxis, majorAxis), angle = cv2.fitEllipse(c)
        
        # detected (contoured) object properties calculation
        perimeter = cv2.arcLength(c,True)
        area_c = area
        hull = cv2.convexHull(c)
        hullArea = cv2.contourArea(hull)
        roundness = 4*np.pi/(hullArea*perimeter)**2
        aspectRatio = w / float(h)
        
        fiberLength = (perimeter-(np.sqrt(perimeter**2)-(16*area)))/4
        fiberWidth = area/fiberLength
        curl = majorAxis/fiberLength
        eccentricity = minorAxis/majorAxis
        solidity = area_c / float(hullArea)
        
        # show the properties
        print("Character no.", OBJ_number+1)
        print("solidity of ", 'OBJ_{}.png'.format(OBJ_number), '='  , solidity)
        print("aspectRatio of ", 'OBJ_{}.png'.format(OBJ_number), '='  , aspectRatio)
        print("roundness of ", 'OBJ_{}.png'.format(OBJ_number), '='  , roundness)
        print("curl of ", 'OBJ_{}.png'.format(OBJ_number), '='  , curl)
        print("eccentricity of ", 'OBJ_{}.png'.format(OBJ_number), '='  , eccentricity)
        print("minorAxis of ", 'OBJ_{}.png'.format(OBJ_number) ," = ", minorAxis)
        print("majorAxis of ", 'OBJ_{}.png'.format(OBJ_number) ," = ", majorAxis)
        
        print ("\n")
        OBJ_number += 1
```

## **3**

เลือก Roundness และ Solidity 


*   Roundness เป็นการวัดความกลมของวัตถุ มีค่าเท่ากับ 1 สำหรับวัตถุทรงกลม และน้อยกว่า 1 สาหรับวัตถุที่แยกออกจากความเป็นวงกลม ยกเว้นว่าจะไม่อ่อนไหวต่อขอบเขตที่ไม่ปกติ


* Solidity เป็นการวัดความหนาแน่นของวัตถุ หรือ การวัดความแข็งแกร่งสามารถวัดได้จากอัตราส่วนของพื้นที่ของวัตถุต่อพื้นที่ของลำตัวนูนของวัตถุ ค่า 1 หมายถึงวัตถุที่เป็นของแข็ง และค่าที่น้อยกว่า 1 จะหมายถึงวัตถุที่มีขอบเขตไม่สม่ำเสมอหรือมีรูอยู่

* สาเหตุที่เลือก 2 ตัวนี้เพราะจะเห็นว่าค่าของ roundness สามารถหาความโค้งกลมของวัตถุได้ดี สังเกตได้จากค่า roundness ของ @ มีค่ามากที่สุดเมื่อเทียบกับทุกตัว และค่า solidity ของ @ ก็จะเห็นได้ว่ามีค่ามากที่สุดเมื่อเทียบกับทุกตัวเนื่องจากเป็นตัวอักษรที่มีขอบเขตสม่ำเสมอและไม่มีรูในตัวอักษรเลย
