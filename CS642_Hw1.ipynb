```python
# Log Transformations
# Open the image.
img_a = cv2.imread('/content/cs642_hw1/Hw1_1A.jpg', 0)

#Display original image
plt.figure(figsize=(10,10))
plt.imshow(img_a, cmap='gray')
plt.colorbar(pad=0.01)
plt.title("Hw1_1A Original Image")
plt.show()
```

```python
print(type(img_a)) # ชนิดออบเจ็กต์ ได้ <class 'numpy.ndarray'>
print(img_a.shape) # รูปร่างอาเรย์
print(img_a.dtype) # ชนิดข้อมูล ได้ uint8
print(img_a.min()) # ค่าต่ำสุด ได้ 0
print(img_a.max()) # ค่าสูงสุด ได้ 255
```

```python
# Apply log transform.
# c = 255/(np.log(1 + np.max(img))) -- ไม่ใช้
c = 15 # กำหนดค่า c = 15
log_transformed = c * (np.log(1 + np.abs(img_a)))

# Specify the data type.
log_transformed = np.array(log_transformed, dtype = np.uint8)

# Save the output.
# cv2.imwrite('log_transformed.jpg', log_transformed)

#Display Log Transform image
# path 
# path = '/content/cs642_hw1/log_transformed.jpg'
# log_transformed_img = cv2.imread(path)
plt.figure(figsize=(10,10))
# plt.imshow(log_transformed_img, cmap='gray')
plt.imshow(log_transformed, cmap='gray')
plt.colorbar(pad=0.01)
plt.title("Hw1_1A apply log transform and c=15")
plt.show()
```

```python
# Apply log transform.
# c = 255/(np.log(1 + np.max(img))) -- ไม่ใช้
c = 40 # กำหนดค่า c = 15
log_transformed = c * (np.log(1 + np.abs(img_a)))

# Specify the data type.
log_transformed = np.array(log_transformed, dtype = np.uint8)

# Save the output.
# cv2.imwrite('log_transformed.jpg', log_transformed)

#Display Log Transform image
# path 
# path = '/content/cs642_hw1/log_transformed.jpg'
# log_transformed_img = cv2.imread(path)
plt.figure(figsize=(10,10))
# plt.imshow(log_transformed_img, cmap='gray')
plt.imshow(log_transformed, cmap='gray')
plt.colorbar(pad=0.01)
plt.title("Hw1_1A apply log transform and c=40")
plt.show()
```

```python
# Open the image.
img_b = cv2.imread('/content/cs642_hw1/Hw1_1B.tif', 0)

#Display original image Hw1_1B
plt.figure(figsize=(10,10))
plt.imshow(img_b, cmap='gray')
plt.colorbar(pad=0.01)
plt.title("Hw1_1B Origianl Image")
plt.show()
```

```python
print(type(img_b)) # ชนิดออบเจ็กต์ ได้ <class 'numpy.ndarray'>
print(img_b.shape) # รูปร่างอาเรย์
print(img_b.dtype) # ชนิดข้อมูล ได้ uint8
print(img_b.min()) # ค่าต่ำสุด ได้ 0
print(img_b.max()) # ค่าสูงสุด ได้ 255
```

```python
# กำหนด gamma = 0.5, c= 15
gamma = 0.5
# Apply gamma correction.
# gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8') -- ไม่ใช้
gamma_corrected = np.array(15 * (img_b**gamma), dtype = 'uint8')
#Display gamma_corrected
plt.figure(figsize=(10,10))
plt.imshow(gamma_corrected, cmap='gray')
plt.colorbar(pad=0.01)
plt.title("Hw1_1B by power transformation/gamma=0.05/c=15")
plt.show()
```

### **ตอบ ข้อที่ 1**
> สมการ A คือ การทำ Log transformation

> สมการ B คือ การทำ Power-Law (Gamma) Transformation โดยที่ 0 < i < 1

> สมการ B คือ การทำ Power-Law (Gamma) Transformation โดยที่ i > 1

> กำหนด c = 15 ซึ่งเป็นค่าคงที่

ภาพ A เป็นการเอา ค่า c คือค่าความสว่างมาคำนวณ ผ่าน log ใน array แล้วบวก 1 ดังสมการ g(x,y) = c * (np.log(1 + np.abs(f(x,y)))) --> ผลลัพธ์ที่ได้ คือ ภาพมีความสว่างเพิ่มขึ้น จากเดิม ดังรูป Hw1_1A apply log transform and c=15 :

ภาพ B เป็นการปรับความสว่างของภาพ โดยการ ทำ Power Law Transformation ดังสมการ g(x,y) = np.array(c * (f(x,y)**gamma), dtype = 'uint8') โดยค่า i หรือ gamma = 0.05 --> ผลลัพธ์ที่ได้ คือ ภาพสว่างจาก ภาพต้นฉบับ ดังรูป Hw1_1B by power transformation/gamma=0.05/c=15

ภาพ C เป็นการปรับความสว่างของภาพคล้ายของ B แต่ จะแตกต่างที่ค่า i หรือ ค่า gamma ซึ่งมีค่ามากกว่า 1 ดังสมการ g(x,y) = np.array(c * (f(x,y)**gamma), dtype = 'uint8') โดยค่า i หรือ gamma = 2 --> ผลลัพธ์ที่ได้ คือ ภาพสว่างจาก ภาพต้นฉบับ ดังรูป Hw1_1C by power transformation/gamma=2/c=15
